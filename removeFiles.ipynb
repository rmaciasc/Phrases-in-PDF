```python
import os
import pandas as pd
```

```python
res = pd.read_csv('./Resumen.csv')
# imgs = pd.read_csv('./ResumenIMGS2.csv')

imgs = res[res['Page'].isnull()].reset_index()
imgsName = imgs['Title'].tolist()
```

```python
dir = os.getcwd()
dir = os.path.join(dir, 'CNDHnotText')
for foldername, subfolders, files in os.walk(os.path.join(dir)):
    for file in files:
        if file not in imgsName:
            print('Removing: ' + file)
            os.remove(os.path.join(foldername, file))
```
